## Mapa logístico

$ x_{n+1}=rx_{n}\left(1-x_{n}\right)$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
npuntos = 2000
xs=[]
rs=[]
nrs = 1000
x = np.zeros(npuntos)
x[0]= 0.1
transitorio = 1000

def genmap(r,nps,trans):
    xps= []
    rs = []
    for i in range(1,nps):
        x[i] = x[i-1]*r*(1-x[i-1]) 
    xps.append(x[trans:nps])
    return(xps)

x26 = genmap(2.6,npuntos,transitorio)
plt.title("r = 2.6")
plt.scatter([range(0,npuntos-transitorio)],x26,s=0.1,color='black')
plt.show()

x35 = genmap(3.5,npuntos,transitorio)
plt.title("r = 3.5")
plt.scatter([range(0,npuntos-transitorio)],x35,s=0.1,color='black')
plt.show()

x38 = genmap(3.8,npuntos,transitorio)
plt.title("r = 3.8")
plt.scatter([range(0,npuntos-transitorio)],x38,s=0.1,color='black')
plt.show()
```

```python
plt.figure(figsize=(14,14))
npuntos = 4000
xs=[]
rs=[]
nrs = 1000
x = np.zeros(npuntos)
x[0]= 0.1

r = np.linspace(2.6,4,nrs)
for j in range(nrs):
    x = np.zeros(npuntos)
    x[0]= 0.1
    for i in range(1,npuntos):
        x[i] = x[i-1]*r[j]*(1-x[i-1]) 
    xs.append(x[transitorio:npuntos])
    rs.append([r[j]]*(npuntos-transitorio))

plt.title("Mapa logístico")
plt.scatter(rs,xs,color='blue',s=0.001, alpha=0.2)
plt.show()   # muestra la gráfica
```
